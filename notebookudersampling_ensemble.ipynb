```python
# Simpan salinan data sebelum outlier removal (untuk visualisasi)
df_before_outlier = df.copy()
import matplotlib.pyplot as plt
import seaborn as sns

# Buat grid layout otomatis sesuai jumlah kolom
n_cols = 4  # jumlah boxplot per baris
n_rows = -(-len(df.columns) // n_cols)  # ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()  # supaya bisa diiterasi

for i, col in enumerate(df.columns):
    sns.boxplot(y=df_before_outlier[col], color='skyblue', ax=axes[i])
    axes[i].set_title(f"{col}")

# Hapus axis kosong kalau jumlah kolom tidak pas
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle(" Boxplot Sebelum Outlier Removal", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
### Outlier Detection and Removal using Z-score
z_scores = np.abs(stats.zscore(df))
df = df[(z_scores < 3).all(axis=1)]  # Keep only rows where all Z-scores < 3
```

```python
print("Shape after removing outliers:", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n Boxplot Setelah Outlier Removal")

# Tentukan layout grid
n_cols = 4  # jumlah boxplot per baris
n_rows = -(-len(df.columns) // n_cols)  # ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()

# Loop setiap kolom dan pasang boxplot ke grid
for i, col in enumerate(df.columns):
    sns.boxplot(y=df[col], color='lightgreen', ax=axes[i])
    axes[i].set_title(col)

# Hapus subplot kosong kalau jumlah kolom tidak pas
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Judul utama
fig.suptitle(" Boxplot Setelah Outlier Removal", fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Memisahkan fitur (X) dan label (y)
X = df.drop('Label', axis=1)
y = df['Label']

print("Shape of features (X):", X.shape)
print("Shape of target (y):", y.shape)
```

```python
print(X.info())
print(y.value_counts())
```

```python
import pandas as pd
## under sampling
# Gabungkan X dan y biar gampang sampling
df_full = X.copy()
df_full["Label"] = y

# Cari jumlah terkecil
min_count = df_full["Label"].value_counts().min()  # 4741

# Undersampling per kelas
df_balanced = (
    df_full.groupby("Label", group_keys=False)
           .apply(lambda x: x.sample(min_count, random_state=42))
)
### data balance

# Pisahkan lagi X dan y
X_balanced = df_balanced.drop("Label", axis=1)
y_balanced = df_balanced["Label"]

print(X_balanced.shape)        # (23705, 11)
print(y_balanced.value_counts())
```

```python
X_balanced.info()
```

## Prepocessing

```python
# Histogram sebelum normalisasi
X_balanced.hist(bins=30, figsize=(15, 10))
plt.suptitle("Histogram fitur Sebelum Normalisasi", fontsize=16)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
# Split data menjadi 80% data train dan 20% data test
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)
```

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()

# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Data Train after normalisation = {X_train_scaled.shape}")
print(f"Data Test after normalisation = {X_test_scaled.shape}")
```

```python
# Histogram setelah normalisasi
X_train_scaled_df = pd.DataFrame(X_train_scaled)
X_train_scaled_df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Histogram Setelah train Normalisasi", fontsize=16)
plt.show()
```

```python
# Inisialisasi PCA tanpa mengurangi dimensi dulu (hanya untuk melihat explained variance)
pca = PCA(n_components=None) # Set n_components to None to keep all components initially
pca.fit(X_train_scaled)

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--', color='b')
plt.xlabel('Jumlah Komponen')
plt.ylabel('Kumulatif Variance yang Dijelaskan')
plt.title('Explained Variance oleh PCA Components')
plt.grid(True)
plt.show()
```

```python
# Tampilkan variansi kumulatif
explained_variance = np.cumsum(pca.explained_variance_ratio_)
for i, var in enumerate(explained_variance):
    print(f"Komponen ke-{i+1}: {var:.4f} variance terjelaskan")

# Misal ingin menyimpan komponen yang menjelaskan >=90% variance
n_components = np.argmax(explained_variance >= 0.90) + 1
print(f"\n Jumlah komponen optimal (90% variance): {n_components}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Buat dictionary model
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": GaussianNB(),
}

# Dictionary untuk simpan hasil evaluasi
results = {}

for name, model in models.items():
    print(f"\n Training {name}...")
    
    # Training pakai data yang sudah discaled
    model.fit(X_train_scaled, y_train)
    
    # Prediction
    y_pred = model.predict(X_test_scaled)
    
    # Evaluasi
    report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
    
    # Simpan hasil utama (macro avg agar balance antar kelas)
    results[name] = {
        "Accuracy": report["accuracy"],
        "Precision": report["macro avg"]["precision"],
        "Recall": report["macro avg"]["recall"],
        "F1-Score": report["macro avg"]["f1-score"]
    }
    
    # Tampilkan laporan lengkap
    print(f"\nClassification Report ({name}):")
    print(classification_report(y_test, y_pred, zero_division=0))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Tampilkan perbandingan hasil dalam bentuk tabel
df_results = pd.DataFrame(results).T
print("\n Perbandingan Model:")
print(df_results)

# Visualisasi perbandingan
df_results.plot(kind="bar", figsize=(10,6))
plt.title("Perbandingan Akurasi, Precision, Recall, dan F1-Score Antar Model")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.legend(loc="lower right")
plt.show()
```

```python
# === Cell 1: Setup & Imports (khusus ENSEMBLE) ===
RANDOM_STATE = 42  # seed global agar hasil stabil di setiap run

# --- Base models (akan dipakai di Voting/Stacking) ---
from sklearn.ensemble import VotingClassifier, StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# --- Metrik evaluasi (untuk test set; tanpa cross-validation) ---
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# --- Imbalanced-learn: ensemble & pipeline ---
from imblearn.ensemble import (
    BalancedRandomForestClassifier,
    BalancedBaggingClassifier,
    EasyEnsembleClassifier,
    RUSBoostClassifier,
)

# Digunakan hanya untuk Voting/Stacking (undersampling eksternal pada TRAIN saja).
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# --- Asumsi penting ---
# Di sel-sel sebelumnya (bagian lama notebook), variabel berikut SUDAH tersedia:
# X_train, X_test, y_train, y_test
# Tidak akan mengubahnya. Untuk ensemble tipe Voting/Stacking,
# Membungkus model dengan Pipeline(RandomUnderSampler -> Model) agar aman dari data leakage.

print("Imports ensemble OK. Ready to add models.")
```

```python
# === Cell 2 (revisi): evaluasi + auto-detect nama variabel split ===
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Cari variabel yang tersedia di globals, pilih yang pertama ketemu
ns = globals()
def pick(*names):
    for n in names:
        if n in ns:
            return ns[n], n
    raise NameError(f"Tidak menemukan salah satu dari: {names}. "
                    "Pastikan sel split train/test sudah dijalankan.")

# Ambil X/y train/test apapun penamaannya (huruf besar/kecil)
XTR, XTR_name = pick('X_train','x_train')
XTE, XTE_name = pick('X_test','x_test')
YTR, YTR_name = pick('y_train','Y_train','train_y','y_tr')
YTE, YTE_name = pick('y_test','Y_test','test_y','y_te')

# Hitung proporsi train/test dari variabel yang berhasil ditemukan
_TOTAL = len(XTR) + len(XTE)
_TRAIN_SIZE = round(len(XTR) / _TOTAL, 1)
_TEST_SIZE  = round(len(XTE) / _TOTAL, 1)

# Kontainer hasil ENSEMBLE (tidak menyentuh tabel lamamu)
df_ensemble = pd.DataFrame(columns=[
    "Model", "Normalisasi", "PCA", "Accuracy", "Precision", "Recall", "F1-Score",
    "Train Size", "Test Size", "Keterangan"
])

def _eval_metrics(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    return acc, prec, rec, f1

def add_result_row(model_name, y_true, y_pred, 
                   normalisasi="Tidak", pca="-", keterangan="-"):
    acc, prec, rec, f1 = _eval_metrics(y_true, y_pred)
    row = {
        "Model": model_name,
        "Normalisasi": normalisasi,
        "PCA": pca,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-Score": f1,
        "Train Size": _TRAIN_SIZE,
        "Test Size": _TEST_SIZE,
        "Keterangan": keterangan
    }
    global df_ensemble
    df_ensemble = pd.concat([df_ensemble, pd.DataFrame([row])], ignore_index=True)

print("Eval helper siap. Dipakai variabel:",
      f"XTR={XTR_name}, XTE={XTE_name}, YTR={YTR_name}, YTE={YTE_name}",
      "| Train/Test size =", _TRAIN_SIZE, "/", _TEST_SIZE)
```

```python
# === Cell 2b (HOTFIX): metrik otomatis binary / multiclass ===
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def _eval_metrics(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    labels = np.unique(np.concatenate([y_true, y_pred]))

    acc = accuracy_score(y_true, y_pred)

    if labels.size == 2:
        # Binary: tentukan positive label secara aman
        if 1 in labels:            pos = 1
        elif "1" in labels.tolist(): pos = "1"
        else:                      pos = sorted(labels.tolist())[-1]  # ambil label terbesar
        prec = precision_score(y_true, y_pred, average="binary", pos_label=pos, zero_division=0)
        rec  = recall_score(y_true, y_pred, average="binary", pos_label=pos, zero_division=0)
        f1   = f1_score(y_true, y_pred, average="binary", pos_label=pos, zero_division=0)
    else:
        # Multiclass: pakai weighted agar aman
        prec = precision_score(y_true, y_pred, average="weighted", zero_division=0)
        rec  = recall_score(y_true, y_pred, average="weighted", zero_division=0)
        f1   = f1_score(y_true, y_pred, average="weighted", zero_division=0)

    return acc, prec, rec, f1
```

```python
# === Cell 3 (revisi): Soft Voting + RUS (TRAIN saja) ===

# Base models
rf  = RandomForestClassifier(n_estimators=300, random_state=RANDOM_STATE)
dt  = DecisionTreeClassifier(random_state=RANDOM_STATE)
knn = KNeighborsClassifier(n_neighbors=5)
nb  = GaussianNB()

# Voting (soft)
voter = VotingClassifier(
    estimators=[('rf', rf), ('dt', dt), ('knn', knn), ('nb', nb)],
    voting='soft'
)

# Pipeline: RandomUnderSampler -> Voting
pipe_vote = Pipeline(steps=[
    ('under', RandomUnderSampler(sampling_strategy='auto', random_state=RANDOM_STATE)),
    ('clf', voter)
])

# Train & eval
pipe_vote.fit(XTR, YTR)
y_pred_vote = pipe_vote.predict(XTE)

add_result_row(
    model_name="VotingClassifier",
    y_true=YTE,
    y_pred=y_pred_vote,
    normalisasi="Tidak",
    pca="-",
    keterangan="Ensemble: Voting + RUS(train)"
)

df_ensemble.tail(1)
```

```python
# === Cell 4: Balanced Ensembles (tanpa undersampling eksternal) ===
# - Fit di TRAIN (XTR, YTR), prediksi di TEST (XTE), lalu tambah ke df_ensemble

import numpy as np

results_info = []

# 1) Balanced Random Forest
brf = BalancedRandomForestClassifier(
    n_estimators=300,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
brf.fit(XTR, YTR)
y_pred_brf = brf.predict(XTE)
add_result_row(
    model_name="BalancedRandomForest",
    y_true=YTE, y_pred=y_pred_brf,
    normalisasi="Tidak", pca="-",
    keterangan="Ensemble: BalancedRF (internal balance)"
)

# 2) Balanced Bagging (base estimator = Decision Tree)
bbag = BalancedBaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=RANDOM_STATE),
    n_estimators=50,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
bbag.fit(XTR, YTR)
y_pred_bbag = bbag.predict(XTE)
add_result_row(
    model_name="BalancedBagging",
    y_true=YTE, y_pred=y_pred_bbag,
    normalisasi="Tidak", pca="-",
    keterangan="Ensemble: BalancedBagging (DT base)"
)

# 3) EasyEnsemble (beberapa AdaBoost pada subset seimbang berbeda)
ees = EasyEnsembleClassifier(
    n_estimators=10,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
ees.fit(XTR, YTR)
y_pred_ees = ees.predict(XTE)
add_result_row(
    model_name="EasyEnsemble",
    y_true=YTE, y_pred=y_pred_ees,
    normalisasi="Tidak", pca="-",
    keterangan="Ensemble: EasyEnsemble (AdaBoost bags)"
)

# 4) RUSBoost (AdaBoost + Random Undersampling tiap iterasi)
rusb = RUSBoostClassifier(
    n_estimators=200,
    learning_rate=1.0,
    random_state=RANDOM_STATE
)
rusb.fit(XTR, YTR)
y_pred_rusb = rusb.predict(XTE)
add_result_row(
    model_name="RUSBoost",
    y_true=YTE, y_pred=y_pred_rusb,
    normalisasi="Tidak", pca="-",
    keterangan="Ensemble: RUSBoost"
)

# Tampilkan ringkasan hasil ensemble (urut F1 menurun)
df_ensemble = df_ensemble.sort_values("F1-Score", ascending=False).reset_index(drop=True)
df_ensemble
```

```python
# === Cell 5 (revisi): XGBoost untuk MULTICLASS (label-encode 0..K-1) ===
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np

# 1) Encode label ke 0..K-1 agar kompatibel XGBoost
le = LabelEncoder()
le.fit(np.unique(np.concatenate([YTR, YTE])))   # aman: pastikan semua kelas ter-cover
ytr_enc = le.transform(YTR)
yte_enc = le.transform(YTE)
n_classes = len(le.classes_)

# 2) Hyperparameter dasar yang stabil
xgb_params = dict(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_lambda=1.0,
    random_state=RANDOM_STATE,
    n_jobs=-1,
)

# 3A) XGB + RUS (TRAIN saja)  multiclass
xgb_rus = XGBClassifier(objective='multi:softmax', num_class=n_classes, **xgb_params)
pipe_xgb_rus = Pipeline(steps=[
    ('under', RandomUnderSampler(sampling_strategy='auto', random_state=RANDOM_STATE)),
    ('clf', xgb_rus)
])
pipe_xgb_rus.fit(XTR, ytr_enc)
y_pred_xgb_rus_enc = pipe_xgb_rus.predict(XTE)
y_pred_xgb_rus = le.inverse_transform(y_pred_xgb_rus_enc)

add_result_row(
    model_name="XGBoost (RUS train)",
    y_true=YTE, y_pred=y_pred_xgb_rus,
    normalisasi="Tidak", pca="-",
    keterangan=f"Ensemble: XGB multiclass + RUS(train), K={n_classes}"
)

# 3B) Varian scale_pos_weight hanya relevan untuk BINARY  skip kalau multiclass
if n_classes == 2:
    pos = 1 if 1 in le.classes_ else le.classes_[-1]
    pos_count = np.sum(ytr_enc == le.transform([pos])[0])
    neg_count = len(ytr_enc) - pos_count
    spw = (neg_count / max(1, pos_count))

    xgb_spw = XGBClassifier(objective='binary:logistic', scale_pos_weight=spw, **xgb_params)
    xgb_spw.fit(XTR, ytr_enc)
    y_pred_xgb_spw = le.inverse_transform(xgb_spw.predict(XTE))

    add_result_row(
        model_name="XGBoost (scale_pos_weight)",
        y_true=YTE, y_pred=y_pred_xgb_spw,
        normalisasi="Tidak", pca="-",
        keterangan=f"Ensemble: XGB binary + spw={spw:.2f}"
    )
else:
    print("Lewati varian scale_pos_weight karena target multiclass (", n_classes, "kelas ).")

# 4) Lihat tabel terbarunya
df_ensemble = df_ensemble.sort_values("F1-Score", ascending=False).reset_index(drop=True)
df_ensemble
```
